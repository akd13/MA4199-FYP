### RNASeq

```python
for idx,t in enumerate(rep1_timepoint_rnaseq):
    df = pd.read_csv(replicate1_rnaseq+t,sep='\t')
    print(df[['txReads']].sum(axis=0).values[0])

print('\n')
for idx,t in enumerate(rep1_timepoint_rnaseq):
    df = pd.read_csv(replicate1_rnaseq+t,sep='\t')
    print(df[['cdReads']].sum(axis=0).values[0])
print('\n')

for idx,t in enumerate(rep1_timepoint_rnaseq):
    df = pd.read_csv(replicate1_rnaseq+t,sep='\t')
    print(df[['cdReads']].count(axis=0).values[0])
print('\n')

for idx,t in enumerate(rep1_timepoint_rnaseq):
    df = pd.read_csv(replicate1_rnaseq+t,sep='\t')
    df_new = df[(df['cdReads'] >= 10)]
    print(df_new[['cdReads']].count(axis=0).values[0])
print('\n')

for idx,t in enumerate(rep1_timepoint_rnaseq):
    df = pd.read_csv(replicate1_rnaseq+t,sep='\t')
    df_new = df[(df['cdReads'] >= 20)]
    print(df_new[['cdReads']].count(axis=0).values[0])
print('\n')
    
for idx,t in enumerate(rep1_timepoint_rnaseq):
    df = pd.read_csv(replicate1_rnaseq+t,sep='\t')
    df_new = df[(df['cdReads'] >= 50)]
    print(df_new[['cdReads']].count(axis=0).values[0])
print('\n')

for idx,t in enumerate(rep1_timepoint_rnaseq):
    df = pd.read_csv(replicate1_rnaseq+t,sep='\t')
    df_new = df[(df['cdReads'] >= 100)]
    print(df_new[['cdReads']].count(axis=0).values[0])
```
