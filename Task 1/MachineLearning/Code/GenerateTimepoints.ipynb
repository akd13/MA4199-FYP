```python
import numpy as np
import pandas as pd
from functools import reduce
```

```python
previous = 'Replicate2/RNASeq_chrE/geneTXCD_RPKMoutput_EVadjusted_ReadOutput_'
timepoints = ['CGATGT-s_5_1_genome.txt','TTAGGC-s_5_1_genome.txt',
             'CAGATC-s_5_1_genome.txt','GCCAAT-s_5_1_genome.txt',
             'CTTGTA-s_5_1_genome.txt'
            ]
dataframes = []
for idx,file in enumerate(timepoints):
    df = pd.read_csv(previous+file, sep="	", na_values=['-'])
    df = df.dropna(how='any')
    df = df[['AccNum', 'GeneName','cdReads',
       'cdRPKM']]
    df.rename(columns={'cdRPKM':'cdRPKM'+str(idx), 'cdReads':'cdReads'+str(idx)}, inplace=True)
    dataframes.append(df)
```

```python
df_rnaseq = reduce(lambda left,right: pd.merge(left,right,on=['AccNum','GeneName']),
                   dataframes)
df_rnaseq.head()
```

```python
df_rnaseq.shape
```

```python
df_rnaseq.to_csv('Replicate2/RNASeq_chrE_mockT0.txt',sep='\t',index=False)
```

```python
previous = 'Replicate2/RPF_chrE/geneTXCD_RPKMoutput_EVadjusted_ReadOutput_'
timepoints = ['CGATGT-s_6_1_genome.txt','ATCACG-s_6_1_genome.txt','TTAGGC-s_6_1_genome.txt',
             'CAGATC-s_6_1_genome.txt','GCCAAT-s_6_1_genome.txt',
             'CTTGTA-s_6_1_genome.txt'
            ]
dataframes = []
for idx,file in enumerate(timepoints):
    df = pd.read_csv(previous+file, sep="	", na_values=['-'])
    df = df.dropna(how='any')
    df = df[['AccNum', 'GeneName','cdReads',
       'cdRPKM']]
    df.rename(columns={'cdRPKM':'cdRPKM'+str(idx), 'cdReads':'cdReads'+str(idx)}, inplace=True)
    dataframes.append(df)
df_rpf = reduce(lambda left,right: pd.merge(left,right,on=['AccNum','GeneName']), dataframes)
df_rpf.head()
```

```python
df_rpf.to_csv('Replicate2/RPF_chrE_mockT0added.txt',sep='\t',index=False)
```

```python
df_rpf.shape
```

```python
df_rnaseq = df_rnaseq[(df_rnaseq['cdReads0'] >= 10) & (df_rnaseq['cdReads1'] >= 10) 
                      & (df_rnaseq['cdReads2'] >= 10)& (df_rnaseq['cdReads3'] >= 10)
                      & (df_rnaseq['cdReads4'] >= 10)]
df_rpf = df_rpf[(df_rpf['cdReads0'] >= 10) 
                & (df_rpf['cdReads1'] >= 10) 
                & (df_rpf['cdReads2'] >= 10)
                & (df_rpf['cdReads3'] >= 10)& (df_rpf['cdReads4'] >= 10)]
```

```python
df_rnaseq.shape
```

```python
df_rpf.shape
```

```python
df_TE = reduce(lambda left,right: pd.merge(left,right,on=['AccNum','GeneName']), [df_rnaseq,df_rpf])
```

```python
df_TE.shape
```

```python
df_TE.head()
```

```python
df_TE[['cdRPKM0_x']] = df_TE[['cdRPKM0_y']].div(df_TE['cdRPKM0_x'].values,axis=0)
df_TE[['cdRPKM1_x']] = df_TE[['cdRPKM1_y']].div(df_TE['cdRPKM1_x'].values,axis=0)
df_TE[['cdRPKM2_x']] = df_TE[['cdRPKM2_y']].div(df_TE['cdRPKM2_x'].values,axis=0)
df_TE[['cdRPKM3_x']] = df_TE[['cdRPKM3_y']].div(df_TE['cdRPKM3_x'].values,axis=0)
df_TE[['cdRPKM4_x']] = df_TE[['cdRPKM4_y']].div(df_TE['cdRPKM4_x'].values,axis=0)

for i in range(0,5):
    df_TE.rename(columns={'cdRPKM'+str(i)+'_x':'TE'+str(i)}, inplace=True)
```
