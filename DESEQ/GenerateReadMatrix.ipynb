```python
import pandas as pd
```

```python
pre ='No_ChrE'
```

```python
files = ['RNASeq_chrE_replicate1.csv','RNASeq_chrE_replicate2.csv',
         'RPF_chrE_replicate1.csv','RPF_chrE_replicate2.csv']
```

```python
df = pd.read_csv(pre+'/'+files[0],sep=',')  
```

```python
for file in files[1:]:
    df_new = pd.read_csv(pre+'/'+file,sep=',')
    df = df.merge(df_new, left_on='AccNum', right_on='AccNum')
```
